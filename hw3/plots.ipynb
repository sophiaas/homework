```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sns.set_style('darkgrid')
from pylab import rcParams
rcParams['figure.figsize'] = 10, 7
```

```python
atari_vanilla = pd.DataFrame.from_dict(pd.read_pickle('logs/atari_10-10-2018_0149.pkl'))
atari_double = pd.DataFrame.from_dict(pd.read_pickle('logs/atari_doubleq_10-10-2018_0308.pkl'))
```

```python
atari_vanilla['model'] = ['vanilla'] * len(atari_vanilla)
atari_double['model'] = ['double q'] * len(atari_double)
```

```python
atari_results = pd.concat([atari_vanilla, atari_double])
```

```python
sns.lineplot(x='t', y='mean_reward', hue='model', data=atari_results[atari_results.model=='vanilla']);
```
